```python
import pandas as pd
```

```python
mel_df = pd.read_csv('/Users/nagarjun/Desktop/PRedictive Analytics/melb_data.csv')
mel_df
mel_df.describe()
```

```python
mel_df.columns
```

```python
mel_df =mel_df.dropna(axis=0)
```

```python
mel_df
```

```python
y=mel_df.Price
y
```

```python
mel_features= ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
x = mel_df[mel_features]
x.describe()
```

```python
from sklearn.tree import DecisionTreeRegressor as DTR
mel_model = DTR(random_state=1)

mel_model.fit(x,y)
```

```python
print('making the prediction for 5 houses:')
print(x.head())
```

```python
print('the predictions are')
print(mel_model.predict(x.head()))
```

```python
#model evaluation
from sklearn.metrics import mean_absolute_error
model_predicted = mel_model.predict(x)
mean_absolute_error(y,model_predicted)
```

```python
from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y=train_test_split(x,y, random_state=0)
mel_model= DTR()
```

```python
mel_model.fit(train_X, train_y)
```

```python
val_pred = mel_model.predict(val_X)
print(mean_absolute_error(val_y, val_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest_model = RandomForestRegressor(random_state=1)
forest_model.fit(train_X, train_y)
mel_pred= forest_model.predict(val_X)
print(mean_absolute_error(val_y,mel_pred))

```
